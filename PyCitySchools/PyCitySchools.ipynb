```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
total_schools = len(school_data)
total_students = len(school_data_complete)
total_budget = school_data["budget"].sum()
avg_math_score = student_data["math_score"].mean()
avg_reading_score = student_data["reading_score"].mean()
math_pass = len(student_data[student_data["math_score"] >= 70])/len(school_data_complete)*100
reading_pass = len(student_data[student_data["reading_score"] >= 70])/len(school_data_complete)*100
overall_pass = len(student_data[(student_data["math_score"] >= 70)&(student_data["reading_score"]>= 70)])/len(school_data_complete)*100
                                                          
df = pd.DataFrame({"Total Schools" : [total_schools], "Total Students": [total_students], "Total Budget": [total_budget], "Average Math Score":[avg_math_score], "Average Reading Score": [avg_reading_score],"% Passing Math":[math_pass],"% Passing Reading":[reading_pass],"% Overall Passing":[overall_pass]})
df.head()
```

```python

school_data_sorted = school_data.rename(columns ={"type":"school type", "size":"Total Students", "budget":"Total School Budget"})
del school_data_sorted["School ID"]
school_data_sorted
```

```python
school_data_sorted = school_data_sorted.sort_values(["school_name"], ascending = True)
school_data_sorted = school_data_sorted.reset_index(drop=False)
school_data_sorted.head(16)
```

```python
x = student_data[student_data["math_score"]>=70].groupby("school_name").size()
y = student_data.groupby("school_name").size()
passing_math_school = x/y*100

a = student_data[student_data["reading_score"]>=70].groupby("school_name").size()
b = student_data.groupby("school_name").size()
passing_reading_school = a/b*100

c = student_data[(student_data["math_score"] >= 70) & (student_data["reading_score"]>=70)].groupby("school_name").size()
d = student_data.groupby("school_name").size()
overall = c/d*100

overall
```

```python
school_data_sorted["Per Student Budget"] = school_data_sorted["Total School Budget"]/school_data_sorted["Total Students"]
school_data_sorted["Average Math Score"] = student_data.groupby("school_name")["math_score"].mean().values
school_data_sorted["Average Reading Score"] = student_data.groupby("school_name")["reading_score"].mean().values

school_data_sorted["% Passing Math"] = passing_math_school.values
school_data_sorted["% Passing Reading"] = passing_reading_school.values
school_data_sorted["% Overall Passing"] = overall.values


school_data_sorted
```

```python
school_data_sorted["Per Student Budget"] = school_data_sorted["Per Student Budget"].astype(int)
school_data_sorted["Per Student Budget"] = school_data_sorted["Per Student Budget"].apply("${:.2f}".format)
school_data_sorted["Total School Budget"] = school_data_sorted["Total School Budget"].astype(int)
school_data_sorted["Total School Budget"] = school_data_sorted["Total School Budget"].apply("${:.2f}".format)
school_data_sorted

#The above code works for formatting 
```

```python

school_data_sorted = school_data_sorted.sort_values(["% Overall Passing"], ascending = False)
school_data_sorted = school_data_sorted.reset_index(drop=True)
school_data_sorted.head()
```

```python
school_data_sorted = school_data_sorted.sort_values(["% Overall Passing"], ascending = True)
school_data_sorted = school_data_sorted.reset_index(drop=True)
school_data_sorted.head()
```
